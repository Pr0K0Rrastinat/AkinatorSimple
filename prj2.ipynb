```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_train=pd.read_csv('list_of_character.csv')
df_train
```

```python
for column in df_train.select_dtypes(include='int').columns:
    df_train[column] = df_train[column].astype(str)

df_train['number of hands'] = df_train['number of hands'].astype(int)
```

```python
columnsToBeConverted=df_train.select_dtypes(exclude=np.number).columns 
for column in columnsToBeConverted: 
    uniqueValues = df_train[column].unique() 
    mapping = { value: index for index, value in enumerate(uniqueValues) } 
    df_train[column]=df_train[column].map(mapping) 
df_train
```

```python
y=df_train['Name']
X=df_train.drop('Name',axis=1)


X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
X_normalized
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text


clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_normalized, y)
tree_rules = export_text(clf, feature_names=df_train.columns.name)
print(tree_rules)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree


feature_importance = clf.feature_importances_

# Define the feature names based on your dataset

plt.figure(figsize=(20, 15))
plot_tree(clf, filled=True, feature_names=df_train.columns)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np

# Assuming you have X_normalized and y defined

X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.3, random_state=42)
y_pred = clf.predict(X_test)

def accuracy(y_pred, y_test):
    return np.sum(y_pred == y_test) / len(y_test)

acc = accuracy(y_pred, y_test)

print(f"Accuracy: {acc}")

# Calculate precision, recall, and F1-score for multiclass classification
precision = precision_score(y_test, y_pred, average='weighted')
print(f"Precision: {precision}")

recall = recall_score(y_test, y_pred, average='weighted')
print(f"Recall: {recall}")

f1 = f1_score(y_test, y_pred, average='weighted')
print(f"F1-score: {f1}")
```
